```python
import gridgeo


gridgeo.__version__
```

```python
url = 'http://thredds.cencoos.org/thredds/dodsC/CA_DAS.nc'

grid = gridgeo.GridGeo(
    url,
    standard_name='sea_water_temperature'
)
```

```python
grid.mesh
```

```python
grid.outline
```

```python
print(f'The grid has {len(grid.polygons)} polygons, showing the first 5.')

grid.geometry[:5]
```

```python
%matplotlib inline
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import numpy as np


fig, ax = plt.subplots(
    figsize=(9, 9),
    subplot_kw={'projection': ccrs.PlateCarree()}
)

x, y = np.meshgrid(grid.x, grid.y)

ax.plot(
    x, y, 'darkgray',
    x.T, y.T, 'darkgray',
    alpha=0.25,
)

ax.coastlines(resolution='10m');
```

```python
import folium


x, y = grid.outline.centroid.xy

m = folium.Map(location=[y[0], x[0]])
folium.GeoJson(grid.outline.__geo_interface__).add_to(m)
m.fit_bounds(m.get_bounds())

m
```
